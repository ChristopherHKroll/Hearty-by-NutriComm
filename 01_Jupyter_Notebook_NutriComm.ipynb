```python
#hypertension classification according to CDC: SBP >=130mmHg OR DBP >=80mmHg 

#calculating the mean of BP measurements
data['SBP'] = data[['BPXSY1', 'BPXSY2', 'BPXSY3', 'BPXSY4']].mean(axis=1)
data['DBP'] = data[['BPXDI1', 'BPXDI2', 'BPXDI3', 'BPXDI4']].mean(axis=1)

#calculating hypertension (1=yes/0=no)
cond_h = [(data['SBP'] >=130) | (data['DBP'] >=80),
         (np.isnan(data['SBP']) == True) & np.isnan(data['DBP']) == True,
         (data['SBP'] < 130) | (data['DBP'] < 80)]

values_h = [1, np.nan, 0]

data['hypertension'] = np.select(cond_h, values_h)
```

```python
#diabetes classification according to CDC: fasting glucose value >=126 mg/dL

#calculating diabetes (1=yes/0=no)
cond_d = [(data['LBXGLU'] >=126),
         (np.isnan(data['LBXGLU']) == True),
         (data['LBXGLU'] < 130)]

values_d = [1, np.nan, 0]

data['diabetes'] = np.select(cond_d, values_d)
```

```python
#dyslipidemia classification according to CDC: total cholesterol >200 mg/dL

#calculating dyslipidemia (1=yes/0=no)
cond_dy = [(data['LBXTC'] >200),
         (np.isnan(data['LBXTC']) == True),
         (data['LBXTC'] <= 200)]

values_dy = [1, np.nan, 0]

data['dyslipidemia'] = np.select(cond_dy, values_dy)
```

```python
# Backend User Input and Prediction Output
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from datetime import datetime
import tensorflow as tf




# Create profile
# Functions
# Function for single select
def single_select(prompt, options, allow_specify=False):
    """
    Allows the user to select an option from a list. If 'allow_specify' is True,
    the user can input their own response for the last option.

    :param prompt: The prompt text to display.
    :param options: A list of options for the user to choose from.
    :param allow_specify: Whether to allow the user to specify their own option.
    :return: The selected option or user-specified input.
    """
    while True:
        # Display the options to the user
        print(prompt)
        for i, option in enumerate(options, start=1):
            print(f"{i}. {option}")
        
        # Try to read and validate user input
        try:
            input_value = int(input("Select an option by entering the corresponding number: "))
            if 1 <= input_value <= len(options):
                return options[input_value - 1]
            elif input_value == len(options) and allow_specify:
                return input(f"Please specify your {options[-1].lower()}: ")
            else:
                print("Please enter a valid number from the list.")
        except ValueError:
            print("Please enter a number.")
            
# Function for validated input
def get_validated_input(prompt, input_type, format=None):
    """
    Prompts the user for input and validates the input type. If a format is provided,
    it also checks if the input matches the expected format.

    :param prompt: The prompt text to display.
    :param input_type: The expected type of the input (e.g., int, float, str).
    :param format: The expected format of the input, if applicable (e.g., date format).
    :return: The user input, converted to the specified type.
    """
    while True:
        user_input = input(prompt)
        try:
            if input_type is int:
                return int(user_input)
            elif input_type is float:
                return float(user_input)
            elif input_type is str:
                if format == 'date':
                    # Validate the date format
                    datetime.strptime(user_input, '%m/%d/%Y')
                return user_input
            else:
                raise ValueError("Invalid input type.")
        except ValueError:
            print(f"Invalid input. Please enter a valid {input_type.__name__}.")
            if format:
                print(f"The expected format is {format}.")

# Function for optional single-select queries
def optional_single_select(prompt, options):
    print(prompt)
    for idx, option in enumerate(options, 1):
        print(f"{idx}. {option}")
    while True:
        try:
            selection = input("Select an option by entering the number (leave empty if optional and you prefer not to say): ")
            if not selection:  # If the input is optional and the user prefers not to say
                return "Prefer not to say"
            selection = int(selection)
            if 1 <= selection <= len(options):
                return options[selection - 1]
            else:
                print("Please enter a number that corresponds to the options.")
        except ValueError:
            print("Please enter a valid number or leave it empty if you prefer not to say.")
            
# Function to handle multi-select queries
def multi_select(prompt, options):
    print(prompt)
    for idx, option in enumerate(options, 1):
        print(f"{idx}. {option}")
    selected_indices = input("Select all that apply by entering the numbers separated by commas (e.g., 1,4,7): ")
    
    # Split the input by commas and convert to set to remove duplicates
    selected_indices_set = set(selected_indices.replace(" ", "").split(","))
    
    # Convert to list of selected options, filtering out any empty strings or invalid numbers
    selected_options = [options[int(idx) - 1] for idx in selected_indices_set if idx.isdigit() and 1 <= int(idx) <= len(options)]
    
    return selected_options

# Mapping-Functions for numeric values of user input
def map_to_numeric(option_list, selected_option):
    return option_list.index(selected_option)



# User data storage
user_data = {}

# User Input analog to Figma design
# User authentication
email = input("Enter your email: ")
password = input("Enter your password: ")
# user_data['Email'] = email_entry.get()
# user_data['Password'] = password_entry.get()

    
# Options for the reasons question
reasons = [
    "I've been advised by my doctor to improve my heart health",
    "I'm worried about my heart health",
    "I want help making healthy choices"
]
user_data['Reason for Joining'] = single_select("\nSelect a reason for joining:", reasons)

# Options for the gender question
genders = [
    "Female",
    "Male",
    "Other gender identity (please specify)",
    "Prefer Not to Say"
]
# Gender Mapping and Saving
genders_numeric = {g:i for i, g in enumerate(genders)}
user_data['Gender'] = map_to_numeric(genders, single_select("\nSelect your gender:", genders, allow_specify=True))

# Questions for height, weight, and date of birth
user_data['Height (cm)'] = get_validated_input("\nEnter your height (cm): ", float)
user_data['Weight (kg)'] = get_validated_input("\nEnter your weight (kg): ", float)
user_data['Date of Birth'] = get_validated_input("\nEnter your date of birth MM/DD/YYYY: ", str, 'date')

# Calculate BMI and add to user data
height_m = user_data['Height (cm)'] / 100  # Convert height from cm to meters
weight_kg = user_data['Weight (kg)']
bmi = weight_kg / (height_m ** 2)
user_data['BMI'] = bmi

# Calculate age based on date of birth and current date and add to user data
date_of_birth_str = user_data['Date of Birth']
date_of_birth = datetime.strptime(date_of_birth_str, '%m/%d/%Y')
current_date = datetime.now()
age = current_date.year - date_of_birth.year - ((current_date.month, current_date.day) < (date_of_birth.month, date_of_birth.day))
user_data['Age'] = age

# Question on diabetes
diabetes_options = [
    "No",
    "Yes",
    "Prefer not to say"
]
user_data['Diabetes'] = single_select("\nDo you have diabetes?", diabetes_options)

# If the user answered 'Yes' to diabetes, ask for the type
if user_data['Diabetes'] == 'Yes':
    diabetes_type_options = [
        "Type 1 diabetes",
        "Type 2 diabetes",
        "Gestational diabetes",
        "Other",
        "I prefer not to say"
    ]
    user_data['Type of Diabetes'] = single_select("\nIf yes, please select the type of diabetes:", diabetes_type_options, allow_specify=True)
else:
    user_data['Type of Diabetes'] = "N/A"


# Question on hypertension
hypertension_options = [
    "No",
    "Yes",
    "Prefer not to say"
]
user_data['Hypertension'] = single_select("\nDo you have hypertension (high blood pressure)?", hypertension_options)

# Question on dyslipidemia or high cholesterol
cholesterol_options = [
    "No",
    "Yes",
    "Prefer not to say"
]
user_data['Cholesterol'] = single_select("\nDo you have dyslipidemia or high cholesterol?", cholesterol_options)

# Question on blood pressure measurement
blood_pressure_options = ["No", "Yes"]
user_data['Recent Blood Pressure Measurement'] = single_select("\nDo you have a recent blood pressure measurement?", blood_pressure_options)

# If the user answered 'Yes' to blood pressure measurement, ask for the value
if user_data['Recent Blood Pressure Measurement'] == 'Yes':
    user_data['Blood Pressure (mmHg)'] = get_validated_input("\nIf yes, please enter (mmHg): ", str)
else:
    user_data['Blood Pressure (mmHg)'] = "N/A"

# Question on race or ethnicity
race_ethnicity_options = [
    "Asian",
    "Black",
    "Mexican",
    "Other Hispanic",
    "Other, including multiracial persons",
    "White",
    "Prefer not to say"
]
# Race or Ethnicity Mapping and Saving
race_ethnicity_numeric = {r:i for i, r in enumerate(race_ethnicity_options)}
user_data['Race or Ethnicity'] = map_to_numeric(race_ethnicity_options, single_select("\nWhich of the following best describes your race or ethnicity?", race_ethnicity_options))

# Question on annual houshold income (optional)
income_input = input("\nPlease provide your annual household income (optional). Leave empty if you prefer not to say: ")
user_data['Annual Household Income'] = income_input if income_input else "N/A"

# Question on user's level of education
education_levels = [
    "Less than 9th grade",
    "9-11th grade (Includes 12th grade with no diploma)",
    "High school graduate/GED or equivalent",
    "College or AA degree",
    "College graduate or above",
    "Don't Know",
    "Prefer not to say"
]
# Level of Education Mapping and Saving
education_levels_numeric = {e:i for i, e in enumerate(education_levels)}
user_data['Level of Education'] = map_to_numeric(education_levels, optional_single_select("\nPlease provide your level of education (optional):", education_levels))

# Question on frequency of fish consumption
fish_frequency = [
    "Never",
    "Rarely (less than once a month)",
    "Sometimes (once a week)",
    "Frequently (2-4 times a week)",
    "Very frequently (5-6 times a week)",
    "Daily (once a day)",
    "With every meal"
]
user_data['Fish Consumption Frequency'] = single_select("\nHow often do you eat items from the following food group?\nFish:", fish_frequency)

# Question on frequenncy of fruit and vegetables consumption
fruit_veg_frequency = [
    "Never",
    "Rarely (less than once a month)",
    "Sometimes (once a week)",
    "Frequently (2-4 times a week)",
    "Very frequently (5-6 times a week)",
    "Daily (once a day)",
    "With every meal"
]
user_data['Fruit and Vegetables Consumption Frequency'] = single_select("\nHow often do you eat items from the following food group?\nFruit and Vegetables:", fruit_veg_frequency)

# Question on frequency of nuts, seeds and legumes consumption
nuts_seeds_legumes_frequency = [
    "Never",
    "Rarely (less than once a month)",
    "Sometimes (once a week)",
    "Frequently (2-4 times a week)",
    "Very frequently (5-6 times a week)",
    "Daily (once a day)",
    "With every meal"
]
user_data['Nuts, Seeds and Legumes Consumption Frequency'] = single_select("\nHow often do you eat items from the following food group?\nNuts and Seeds:", nuts_seeds_legumes_frequency)

# Question on frequency of processed meat consumption
processed_meat_frequency = [
    "Never",
    "Rarely (less than once a month)",
    "Sometimes (once a week)",
    "Frequently (2-4 times a week)",
    "Very frequently (5-6 times a week)",
    "Daily (once a day)",
    "With every meal"
]
user_data['Processed Meat Consumption Frequency'] = single_select("\nHow often do you eat items from the following food group?\nProcessed Meat:", processed_meat_frequency)

# Question on frequency of red meat consumption
red_meat_frequency = [
    "Never",
    "Rarely (less than once a month)",
    "Sometimes (once a week)",
    "Frequently (2-4 times a week)",
    "Very frequently (5-6 times a week)",
    "Daily (once a day)",
    "With every meal"
]
user_data['Red Meat Consumption Frequency'] = single_select("\nHow often do you eat items from the following food group?\nRed meat:", red_meat_frequency)

# Question on frequency of sugar-sweetened beverages consumption
ssb_frequency = [
    "Never",
    "Rarely (less than once a month)",
    "Sometimes (once a week)",
    "Frequently (2-4 times a week)",
    "Very frequently (5-6 times a week)",
    "Daily (once a day)",
    "With every meal"
]
user_data['SSB Consumption Frequency'] = single_select("\nHow often do you eat items from the following food group?\nSugar-sweetened beverages:", ssb_frequency)

# Question on frequency of whole grains consumption
whole_grains_frequency = [
    "Never",
    "Rarely (less than once a month)",
    "Sometimes (once a week)",
    "Frequently (2-4 times a week)",
    "Very frequently (5-6 times a week)",
    "Daily (once a day)",
    "With every meal"
]
user_data['Whole Grains Consumption Frequency'] = single_select("\nHow often do you eat items from the following food group?\nWhole grains:", whole_grains_frequency)


# Question on dietary requirements or preferences
dietary_options = [
    "Omnivore",
    "Vegetarian",
    "Vegan",
    "Pescatarian",
    "Flexitarian",
    "Ketogenic",
    "Paleo",
    "Gluten-Free",
    "Lactose-Free",
    "Halal",
    "Kosher",
    "Raw Food Diet",
    "Low-Carb Diet",
    "Mediterranean"
]
user_data['Dietary Preferences'] = multi_select("\nWhich of the following best describe your dietary requirements or preferences? (Select all that apply)", dietary_options)


# Additional user input not in the Figma design but for the purpose of the simulation of the prediction model
# Question for marital status
marital_status_options =  [
    "Married",
    "Widowed",
    "Divorced",
    "Separated",
    "Never married",
    "Living with partner",
    "Prefer not to say"
]
# Marital Status Mapping and Saving
marital_status_numeric = {m:i for i, m in enumerate(marital_status_options)}
user_data['Marital Status'] = map_to_numeric(marital_status_options, single_select("\nWhich of the following best describes your marital status?", marital_status_options))

# Convert user data dictionary to DataFrame
# df_new_sample = pd.DataFrame([user_data])

# Create new DataFrame if it does not exist yet
df_data_new = pd.DataFrame() if 'df_data_new' not in globals() else df_data_new

# Cave: Import original Dataframe from CSV file
# df_data_new = pd.read_csv("./data_050224.csv")

# Add sample to the new DataFrame
# df_data_new = pd.concat([df_data_new, df_new_sample], ignore_index=True)
# Saving updated DataFrame
# df_data_new.to_csv('./df_data_new.csv')

# Create mock user sample DataFrame for simulation of the prediction model
# Input of mock demographic and dietary user data
# NOTE: Nutrient Calculation Feature Omission
# Due to current limitations in resources and capabilities,
# the implementation of the nutrient calculation based on food consumption frequencies
# has been omitted from this project phase.
# The envisioned process involves:
# 1. Calculating average food consumption in grams for different user profiles.
# 2. Linking food items with a food composition table via API to obtain composition data.
# 3. Computing nutrient intake from the consumed foods according to user input using the composition data.
# This feature could significantly enhance user experience by providing detailed nutritional insights.
# Future development could consider implementing this process to fully realize the application's potential.

df_mock_sample = pd.DataFrame({
    'fish': 533.75,
    'fruit_veg': 315.0,
    'nuts_seed_legume': 612.81,
    'processed_meat' : 0.0,
    'red_meat': 110.0,
    'ssb': 0.0,
    'whole_grain' : 210.0,
    'DRTKCAL' : 1625.0,
    'DRTPROT':73.91,
    'DRTCARB':222.975,
    'DRTSUGR':51.0,
    'DRTFIBE': 24.65,
    'DRTTFAT': 51.895,
    'DRTSFAT': 11.1520,
    'DRTMFAT': 22.7145,
    'DRTPFAT' : 12.8855,
    'DRTCHOL' : 113.5,
    'RIDAGEYR': [user_data['Age']],
    'RIDRETH1': [user_data['Race or Ethnicity']],
    'DMDEDUC2': [user_data['Level of Education']],
    'RIAGENDR': [user_data['Gender']],
    'DMDMARTL': [user_data['Marital Status']],
    'INDHHIN2': [user_data['Annual Household Income']],
    'SBP': 122.0,
    'DBP':7.06,
    'BMXBMI': [user_data['BMI']],
    'LBDHDD': 42.0,
    'LBXTR': 116.0,
    'LBDLDL': 56.0,
    'LBXTC':121.0
})

# Change data type for models
df_mock_sample = df_mock_sample.astype('float32')

# Convert Pandas DataFrame to Tensor Tensor
mock_input_tensor = tf.convert_to_tensor(df_mock_sample.values, dtype=tf.float32)

# Create new DataFrame if it does not exist yet
df_data_new = pd.DataFrame() if 'df_data_new' not in globals() else df_data_new

# Add mock user sample to the new DataFrame
df_data_new = pd.concat([df_data_new, df_mock_sample], ignore_index=True)

#Saving new  datasets
df_mock_sample.to_csv('./df_mock_sample.csv')
df_data_new.to_csv('./df_data_new.csv')

# Printing head of new datasets
print(df_mock_sample.head())
print(df_data_new.head())


# Make predictions on the mock user sample data
# 1. Logistic Regression Model: Classification and Risk Assessment (Probability)
mock_pred_lr = model.predict(df_mock_sample)
mock_prob_lr = model.predict_proba(df_mock_sample)

# 2. # Neural Network Model for CVD Prediction: Risk Assessment (Probability)
mock_prob_nn_CVD = model_nn_CVD.predict(mock_input_tensor)


# Print Predictions: Classification and Risk Assessment (Probability)
# For the Logistic Regression Model prediction
predicted_class_numeric = int(mock_pred_lr[0])
if predicted_class_numeric == 0:
    prediction_message_lr = "Low risk of cardiovascular disease"
else:
    prediction_message_lr = "High risk of cardiovascular disease"
print("\nRisk assessment by Logistic Regression Model:", prediction_message_lr)

# Accessing the probability of Class 1 for the mock user sample and explaining it
predicted_prob_numeric = float(mock_prob_lr[0][1]) 
print("\nYour estimated risk of cardiovascular disease by Logistic Regression Model is:", np.round(predicted_prob_numeric * 100, decimals=2), "%")

# For the Neural Network Model prediction
# Accessing the probability of the positive class for the mock user sample and explaining it
prob_pos_class_nn = mock_prob_nn_CVD[0][0]  # Assuming there's only one output for the probability of Class 1
if prob_pos_class_nn >= 0.5:
    prediction_message_nn = "High risk of cardiovascular disease"
else:
    prediction_message_nn = "Low risk of cardiovascular disease"
print("\nRisk assessment by Neural Network Model:", prediction_message_nn)
print("\nYour estimated risk of cardiovascular disease by Neural Network Model is:", np.round(prob_pos_class_nn * 100, decimals=2), "%")


```
